```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from numpy.polynomial.polynomial import polyfit
```

```python
plt.scatter(df.TEFF,df.J0395-df.iSDSS)

b,m = polyfit (df.TEFF,df.J0395-df.iSDSS,1)
np.corrcoef(df.TEFF,df.J0395-df.iSDSS)[0,1]
print("y=",b,"+",m,"*x")
plt.plot(df.TEFF,b+m*df.TEFF)
a=np.corrcoef(df.TEFF,df.J0395-df.iSDSS)[0,1]
print(a**2)
```
